```python
import pandas as pd
```

```python
df = pd.read_csv('McDonalds_Financial_Statements.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Market cap ($B)'], marker='o')
plt.title('Market Cap Over Time')
plt.xlabel('Year')
plt.ylabel('Market Cap ($B)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Revenue ($B)'], marker='o', label='Revenue ($B)')
plt.plot(df['Year'], df['Earnings ($B)'], marker='o', label='Earnings ($B)')
plt.title('Revenue vs. Earnings Over Time')
plt.xlabel('Year')
plt.ylabel('Amount ($B)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['Current Ratio'] = df['Cash on Hand ($B)'] / (df['Total liabilities ($B)'] - df['Total debt ($B)'])
df['Quick Ratio'] = (df['Cash on Hand ($B)'] + df['Revenue ($B)']) / (df['Total liabilities ($B)'] - df['Total debt ($B)'])
df['Debt to Equity Ratio'] = df['Total debt ($B)'] / df['Net assets ($B)']
df['Return on Assets (ROA)'] = df['Earnings ($B)'] / df['Total assets ($B)']
df['Return on Equity (ROE)'] = df['Earnings ($B)'] / df['Net assets ($B)']
df['Operating Margin'] = df['Operating Margin (%)'] / 100
df['P/E Ratio'] = df['P/E ratio']
df['P/S Ratio'] = df['P/S ratio']
df['P/B Ratio'] = df['P/B ratio']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Debt to Equity Ratio'], marker='o')
plt.title('Debt to Equity Ratio Over Time')
plt.xlabel('Year')
plt.ylabel('Debt to Equity Ratio')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Return on Assets (ROA)'], marker='o', label='ROA')
plt.plot(df['Year'], df['Return on Equity (ROE)'], marker='o', label='ROE')
plt.title('Return on Assets (ROA) vs Return on Equity (ROE) Over Time')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
plt.plot(df['Year'], df['Current Ratio'], marker='o', label='Current Ratio')
plt.plot(df['Year'], df['Quick Ratio'], marker='o', label='Quick Ratio')
plt.plot(df['Year'], df['Return on Equity (ROE)'] * 100, marker='o', label='ROE (%)')
plt.plot(df['Year'], df['Return on Assets (ROA)'] * 100, marker='o', label='ROA (%)')
plt.title('Financial Ratios Over Time')
plt.xlabel('Year')
plt.ylabel('Ratio / Percentage')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Revenue ($B)', y='Earnings ($B)', hue='Year', data=df, palette='viridis', s=100)
plt.title('Revenue vs Earnings Scatter Plot')
plt.xlabel('Revenue ($B)')
plt.ylabel('Earnings ($B)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['P/E ratio'], kde=True, bins=20)
plt.title('Distribution of P/E Ratio')
plt.xlabel('P/E Ratio')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

CORRELATION HEATMAP

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Financial Metrics')
plt.show()
```

SWOT ANALYSIS

```python
average_revenue_growth = (df['Revenue ($B)'].iloc[-1] - df['Revenue ($B)'].iloc[0]) / df['Revenue ($B)'].iloc[0] * 100
average_operating_margin = df['Operating Margin (%)'].mean()
average_ROE = df['Return on Equity (ROE)'].mean() * 100

print(f"Strengths: Average revenue growth of {average_revenue_growth:.2f}%, Average operating margin of {average_operating_margin:.2f}%, Positive ROE of {average_ROE:.2f}%")
```

Financial Risks

```python
market_volatility = df['Market cap ($B)'].std() / df['Market cap ($B)'].mean() * 100
credit_risk = df['Debt to Equity Ratio'].mean()
operational_risk = df['Operating Margin (%)'].std()

print(f"Market Volatility: {market_volatility:.2f}%")
print(f"Credit Risk: {credit_risk:.2f}")
print(f"Operational Risk: {operational_risk:.2f}")
```

```python
recommendations = [
    "Optimize operational costs and implement cost-control measures.",
    "Explore new markets and increase product offerings.",
    "Invest in high-growth areas and enhance technological capabilities."
]
for rec in recommendations:
    print(rec)
```

```python
import xlsxwriter

workbook = xlsxwriter.Workbook('Financial_Health_Assessment_Report.xlsx')
worksheet = workbook.add_worksheet()

bold_format = workbook.add_format({'bold': True})
header_format = workbook.add_format({'bold': True, 'font_color': 'blue', 'align': 'center'})
text_wrap_format = workbook.add_format({'text_wrap': True})

worksheet.merge_range('B1:E1', 'Financial Health Assessment Report', header_format)

worksheet.write('D2', 'SWOT Analysis', bold_format)
worksheet.write('B3', 'Strengths', bold_format)
worksheet.write('B4', f"1. Average revenue growth of {average_revenue_growth:.2f}%", text_wrap_format)
worksheet.write('B5', f"2. Strong average operating margins of {average_operating_margin:.2f}%", text_wrap_format)
worksheet.write('B6', f"3. Positive average Return on Equity (ROE) of {average_ROE:.2f}%", text_wrap_format)

worksheet.write('C3', 'Weaknesses', bold_format)
worksheet.write('C4', "1. High debt levels compared to equity.", text_wrap_format)
worksheet.write('C5', "2. Fluctuating earnings over the period.", text_wrap_format)
worksheet.write('C6', "3. Low or inconsistent dividend yield.", text_wrap_format)

worksheet.write('D3', 'Opportunities', bold_format)
worksheet.write('D4', "1. Market expansion opportunities identified from revenue growth trends.", text_wrap_format)
worksheet.write('D5', "2. Potential cost reductions based on operating margin analysis.", text_wrap_format)
worksheet.write('D6', "3. Investment in new technologies and product lines.", text_wrap_format)

worksheet.write('E3', 'Threats', bold_format)
worksheet.write('E4', "1. Market volatility reflected in fluctuating market cap and earnings.", text_wrap_format)
worksheet.write('E5', "2. Increased competition impacting revenue growth.", text_wrap_format)
worksheet.write('E6', "3. Economic factors influencing overall financial stability.", text_wrap_format)

worksheet.write('F3', 'Financial Risk Evaluation', bold_format)
worksheet.write('F4', "Market Volatility: {:.2f}%".format(df['Market cap ($B)'].std() / df['Market cap ($B)'].mean() * 100), text_wrap_format)
worksheet.write('F5', "Credit Risk: {:.2f}".format(df['Debt to Equity Ratio'].mean()), text_wrap_format)
worksheet.write('F6', "Operational Risk: {:.2f}".format(df['Operating Margin (%)'].std()), text_wrap_format)

worksheet.write('G3', 'Strategic Recommendations', bold_format)
worksheet.write('G4', "1. Optimize operational costs and implement cost-control measures.", text_wrap_format)
worksheet.write('G5', "2. Explore new markets and increase product offerings.", text_wrap_format)
worksheet.write('G6', "3. Invest in high-growth areas and enhance technological capabilities.", text_wrap_format)
workbook.close()
```
